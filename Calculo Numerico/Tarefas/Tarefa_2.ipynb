EXERCÍCIO 1

Uma TV LED 32” Samsung vendida online nas Americanas custa, a
vista, R\$ 1.305,00 (dados de julho de 2022, pagamento no boleto). Na
opção a prazo são uma entrada de R\$ 139,65 e 11 parcelas mensais de
R$ 139,56. Qual a taxa de juros mensal e anual embutida no plano de
pagamento a prazo?

Utilizando os cálculos de fluxo de caixa, temos que: $$V_{àvista} = V_{entrada} + \sum_{j=1}^{n}\frac{V_{parcelas}}{(1+i)^j}$$

Onde $j$ representa os meses, $n$ o número de parcelas, e $i$ a taxa de juros.

Portanto, para encontrarmos o valor da taxa de juros correspondente, basta subtrairmos $V_{àvista}$ de ambos os lados, encontrando a raiz da função gerada. Para realizar este processo, foi utilizado a aproximação pelo Método da Bissecção, como visto abaixo:

```python
a_vista = 1305    #Valor do produto à vista
entrada = 139.65    #Valor da entrada
parcelas = 139.56    #Valor das parcelas
meses = 11     #Quantos meses parcelados
VP = entrada - a_vista    #Valor presente (Negativo devido a dívida)

def mensal(i):
  z = 0     #Variável acumuladora
  for j in range(meses-10):   #Contador até o mês 1
    y = z + parcelas/pow((1+i),(j+1))     #Variável que irá possuir os termos dependentes da função
    z = y     #Cálculo para realizar o somatório

  return VP + y     #Retorna a função que possui os termos independentes e dependentes (Função geral)

def anual(i):
  z = 0    #Variável acumuladora
  for j in range(meses):     #Contador até os meses correspondentes as parcelas
    y = z + parcelas/pow((1+i),(j+1))     #Variável que irá possuir os termos dependentes da função
    z = y     #Cálculo para realizar o somatório

  return VP + y    #Retorna a função que possui os termos independentes e dependentes (Função geral)

a1 = -0.8    #Ponto A para função mensal
b1 = -0.9    #Ponto B para função mensal
N1 = 50      #Número de iterações para função mensal
a2 = 0       #Ponto A para função anual
b2 = 1       #Ponto B para função anual
N2 = 50      #Número de iterações para função anual

for i in range(N1):    #Contador para realizar a quantidade de iterações N1
  if mensal(a1)*mensal(b1) < 0:   #Verificador
    m1 = (a1 + b1)/2      #Cálculo do ponto médio
    a1 = m1       #Substitui o intervalo [A1;B1] por [M1;B1]

    if mensal(a1)*mensal(b1) > 0:    #Verificador
      a1 = 2*m1 - b1    #Retorna o intervalo [M1;B1] para [A1;B1]
      b1 = m1     #Substitui o intervalo [A1;B1] para [A1;M1]
    
    else:     #Condição para manter as alterações feitas caso o primeiro verificador se mantenha verdadeiro
      continue

for i in range(N2):   #Mesmo código, considerar os comentários anteriores
  if anual(a2)*anual(b2) < 0:
    m2 = (a2 + b2)/2
    a2 = m2

    if anual(a2)*anual(b2) > 0:
      a2 = 2*m2 - b2
      b2 = m2
    
    else:
      continue

print(f'O valor da taxa de juros mensal é de, aproximadamente, {a1*100 :.2f}%\n')    #Mostrar o valor da raiz, isto é, a taxa de juros mensal, com duas casas decimais
print(f'O valor da taxa de juros anual é de, aproximadamente, {a2*100 :.2f}%\n')     #Mostrar o valor da raiz, isto é, a taxa de juros anual, com duas casas decimais
print('Abaixo estão os gráficos das funções mensal e anual: \n')

x = np.linspace(0,5,200)   #Geração de 200 pontos entre -2 e 2 para plotagem do gráfico
fig, ax = plt.subplots()    #Define que o gráfico terá um subplot
ax.plot(x, mensal(x), label = 'Função mensal')   #Plotar a função mensal
ax.plot(x, anual(x), label = 'Função anual')     #Plotar a função anual
ax.set_xlabel('x',fontsize=14)   #Legenda do eixo X
ax.set_ylabel('y',fontsize=14)   #Legenda do eixo Y
ax.set_title('Comparação entre a função mensal e anual',fontsize=14) #Título
ax.legend(loc='best')   #Gera a legenda e dá a posição da mesma
ax.grid(True)    #Coloca o grid de linhas 
plt.show()   #Gera o gráfico
```

EXERCÍCIO 2

Um empreendedor tem a opção de comprar um máquina pelo valor a
vista de R\$129.000,00, mas para isso ele terá que emprestar dinheiro sob
uma taxa de 2% ao mês. Outra opção é parcelar o valor da máquina
em uma entrada de R\$27.000,00 mais quatro parcelas mensais de valores
R\$24.000,00, R\$30.000,00, R\$32.000,00 e R\$22.000,00. Qual a taxa de
juros embutida no pagamento parcelado? Qual dessas opções é melhor
em termos do juros sendo cobrado?

Utilizando os cálculos de fluxo de caixa, temos que: $$V_{àvista} = V_{entrada} + \sum_{j=1}^{n}\frac{V_{parcelas}}{(1+i)^j}$$

Onde $j$ representa os meses, $n$ o número de parcelas, e $i$ a taxa de juros.

Portanto, para encontrarmos o valor da taxa de juros correspondente, basta subtrairmos $V_{àvista}$ de ambos os lados, encontrando a raiz da função gerada. Para realizar este processo, foi utilizado a aproximação pelo Método da Bissecção, como visto abaixo:

```python
a_vista = 129000
i1 = 2/100

entrada = 27000
parcelas = [24000, 30000, 32000, 22000]
VP = entrada - a_vista    #Valor presente, isto é, o valor resultante

def emprestimo(m):  
  return a_vista*(pow((1+i1),m))   #Retorna a função empréstimo

def parcelado(i):
  z = 0    #Variável acumuladora
  for j in range(len(parcelas)):    #Contador até os meses correspondentes as parcelas
    y = z + parcelas[j]/pow((1+i),(j+1))    #Variável que irá possuir os termos dependentes da função
    z = y    #Cálculo para realizar o somatório
 
  return VP + y    #Retorna a função

a1 = 0
b1 = 1
N1 = 50

for i in range(N1):    #Contador para realizar a quantidade de iterações N1
  if parcelado(a1)*parcelado(b1) < 0:     #Verificador
    m1 = (a1 + b1)/2     #Cálculo do ponto médio
    a1 = m1      #Substitui o intervalo [A1;B1] por [M1;B1]

    if parcelado(a1)*parcelado(b1) > 0:          #Verificador
      a1 = 2*m1 - b1         #Retorna o intervalo [M1;B1] para [A1;B1]
      b1 = m1          #Substitui o intervalo [A1;B1] para [A1;M1]
    
    else:
      continue         #Condição para manter as alterações feitas caso o primeiro verificador se mantenha verdadeiro

print(f'A taxa de juros para o pagamento parcelado é de, aproximadamente, {a1*100 :.2f}%. \n')
print(f'Como a taxa de juros para o pagamento parcelado foi maior que a taxa do empréstimo, conclui-se que a melhor opção em termos da taxa de juros é o empréstimo. \n')
print(f'Abaixo estão os gráficos da função referente ao empréstimo e pagamento parcelado: \n')

x = np.linspace(0,5,200)   #Geração de 200 pontos entre -2 e 2 para plotagem do gráfico
fig, ax = plt.subplots()    #Define que o gráfico terá um subplot
ax.plot(x, emprestimo(x), label = 'Função empréstimo')   #Plota a função empréstimo
ax.plot(x, parcelado(x), label = 'Função parcelado')  #Plota a função parcelado
ax.set_xlabel('x',fontsize=14)   #Legenda do eixo X
ax.set_ylabel('y',fontsize=14)   #Legenda do eixo Y
ax.set_title('Comparação entre empréstimo e parcelado',fontsize=14) #Título
ax.legend(loc='best')   #Gera a legenda e dá a posição da mesma
ax.grid(True)    #Coloca o grid de linhas 
plt.show()   #Gera o gráfico

print('\nVale ressaltar que as funções empréstimo e parcelado possuem variáveis diferentes, tendo que analisá-las individualmente.')
```

o investidor com a promessa dos seguintes pagamentos: O pagamento
de parcelas semestrais de R\$ 48,81 a serem pagas nos meses de janeiro

neste título?

No entanto, diferentemente das questões anteriores, este exercício apresenta $V_{entrada} = 0$. Ademais, para realizar este processo, foi utilizado a aproximação pelo Método da Bissecção, como visto abaixo:


```python
valor = 889.18
parcelas = 48.81
mes_atual = 9
anos = 10
meses_total = 12*anos + (12 - mes_atual) + 1    #Quantidade total de meses

def mensal(i):   #Função mensal

  z = 0    #Variável acumuladora
  pagamento_final = 1000/pow((1+i),(meses_total))          #Pagamento final no último mês

  for j in range(1):     #Contador até os meses correspondentes as parcelas (Inicia em 4, conta até o número de meses, com passo de 6)
    y = z + parcelas/pow((1+i),(j+1))     #Variável que irá possuir os termos dependentes da função
    z = y     #Cálculo para realizar o somatório

  return -valor + y + pagamento_final

def anual(i):    #Função anual

  z = 0    #Variável acumuladora
  pagamento_final = 1000/pow((1+i),(meses_total))

  for j in range(12):     #Contador até os meses correspondentes as parcelas (Inicia em 4, conta até o número de meses, com passo de 6)
    y = z + parcelas/pow((1+i),(j+1))     #Variável que irá possuir os termos dependentes da função
    z = y     #Cálculo para realizar o somatório

  return -valor + y + pagamento_final

a1 = 0    #Ponto A para função mensal
b1 = 1  #Ponto B para função mensal
N1 = 50      #Número de iterações para função mensal
a2 = 0       #Ponto A para função anual
b2 = 1       #Ponto B para função anual
N2 = 50      #Número de iterações para função anual

for i in range(N1):    #Contador para realizar a quantidade de iterações N1

  if mensal(a1)*mensal(b1) < 0:   #Verificador

    m1 = (a1 + b1)/2      #Cálculo do ponto médio
    a1 = m1       #Substitui o intervalo [A1;B1] por [M1;B1]

    if mensal(a1)*mensal(b1) > 0:    #Verificador

      a1 = 2*m1 - b1    #Retorna o intervalo [M1;B1] para [A1;B1]
      b1 = m1     #Substitui o intervalo [A1;B1] para [A1;M1]
    
    else:     #Condição para manter as alterações feitas caso o primeiro verificador se mantenha verdadeiro

      continue

for i in range(N2):   #Mesmo código, considerar os comentários anteriores

  if anual(a2)*anual(b2) < 0:

    m2 = (a2 + b2)/2
    a2 = m2

    if anual(a2)*anual(b2) > 0:

      a2 = 2*m2 - b2
      b2 = m2
    
    else:

      continue

print(f'O valor da taxa de juros mensal é de, aproximadamente, {a1*100 :.2f}%\n')    #Mostrar o valor da raiz, isto é, a taxa de juros mensal, com duas casas decimais
print(f'O valor da taxa de juros anual é de, aproximadamente, {a2*100 :.2f}%\n')     #Mostrar o valor da raiz, isto é, a taxa de juros anual, com duas casas decimais
print('Abaixo estão os gráficos das funções geral, mensal e anual: \n')

x = np.linspace(0,1,200)   #Geração de 200 pontos entre -2 e 2 para plotagem do gráfico
fig, ax = plt.subplots()    #Define que o gráfico terá um subplot
ax.plot(x, mensal(x), label = 'Função mensal')   #Plota a função mensal
ax.plot(x, anual(x), label = 'Função anual')    #Plota a função anual
ax.set_xlabel('x',fontsize=14)   #Legenda do eixo X
ax.set_ylabel('y',fontsize=14)   #Legenda do eixo Y
ax.set_title('Comparação entre empréstimo e parcelado',fontsize=14) #Título
ax.legend(loc='best')   #Gera a legenda e dá a posição da mesma
ax.grid(True)    #Coloca o grid de linhas 
plt.show()   #Gera o gráfico
```
