```python
# MÉTODO DE NEWTON (COMUM)

def f(x):
  return x**3 - 0.955712*x**2 - 14.400994*x + 23.142353

def df(x):
  return 3*x**2 - 2*0.955712*x - 14.400994

prec = 0.00000001
x0 = -4
 
for i in range(40):
  x = x0 - (f(x0)/df(x0))
  x0 = x

  print(f'x{i} = {x}')
```

```python
# MÉTODO DE NEWTON (DERIVADA SIMBÓLICA)

x = sp.Symbol('x')

f = x**3 - 0.955712*x**2 - 14.400994*x + 23.142353

dfdx = sp.diff(f,x)

f_numer = sp.lambdify(x,f)
dfdx_numer = sp.lambdify(x,dfdx)

prec = 0.00000001
x0 = -4
y = 0
 
for i in range(40):
  x = x0 - (f_numer(x0)/dfdx_numer(x0))
  x0 = x
  y = x

  print(f'x{i} = {x}')
```
