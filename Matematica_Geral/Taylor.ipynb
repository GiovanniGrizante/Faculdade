Considerando a função abaixo 

\begin{align}
  f(x) = \frac{x}{\sqrt{4x^2+1}}
\end{align}

e o polinômio de Taylor de ordem 2 expandido ao redor de $x_0=1$. O objetivo é observar como o erro decai a media que 
$x \rightarrow x_0$.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def f(x):  #a função
  y = x/np.sqrt(4*x**2+1)
  return y

def df(x): #sua derivada 
  y = 1/np.power(4*x**2+1,3/2)
  return y

def d2f(x): # a segunda derivada
  y = x*((48*x**2)/(4*x**2 + 1)**(5/2) - 4/(4*x**2 + 1)**(3/2)) - (8*x)/(4*x**2 + 1)**(3/2)
  return y

x0 = 1   #ponto de expansão do polinômio de Taylor
def Poli_Taylor_ord_2(x):  #polinomio de Taylor de ordem 2
  y = f(x0)+ df(x0)*(x-x0)+(d2f(x0)/2)*(x-x0)**2
  return y
```

Este aqui é o gráfico que mosta a função e seu polinômio.

```python
x=np.linspace(-2,4,100) #gera um vetor x com 100 valores igualmente espaçados entre -2 até 100
cm = 1/2.54  # centímetos em uma polegada (intch)
fig, ax = plt.subplots(figsize=(15*cm, 10*cm)) #tamanho
ax.plot(x, f(x), label = 'a função f')         #plota f e dá sua respectiva legenda.
ax.plot(x, Poli_Taylor_ord_2(x), label = 'Poli. Taylor ordem 2') 
ax.scatter(x0, f(x0), c='tab:red' , s=30, label='ponto', marker = 'o') #ponto vermelho
ax.set_xlabel('x',fontsize=14)   #legenda dos eixos
ax.set_ylabel('y',fontsize=14)
ax.set_title('f(x) e o poinomio de Taylor',fontsize=14) #titulo
ax.legend(loc='lower right')      #gera a legenda e dá a posição
ax.grid(True)                     #coloca o grid de linhas 
plt.show()                        #mostra 
```

Este aqui é um zoom mostrando onde o polinômio ajusta bem à função.

```python
x=np.linspace(0.7,1.3,100) #gera um vetor x com 100 valores igualmente espaçados de -2 até 100
cm = 1/2.54  # centimeters in inches
fig, ax = plt.subplots(figsize=(15*cm, 10*cm))
ax.plot(x, f(x), label = 'a função')
ax.plot(x, Poli_Taylor_ord_2(x), label = 'Poli. Taylor de ordem 2')
ax.scatter(x0, f(x0), c='tab:red' , s=30, label='ponto', marker = 'o')
ax.set_xlabel('x',fontsize=14)
ax.set_ylabel('y',fontsize=14)
ax.set_title('f(x) e o poinomio de Taylor',fontsize=14)
ax.legend(loc='lower right') 
ax.grid(True)
plt.show()
```

Este gráfico mostra a diferença 

$\displaystyle g(x) = f(x) - P_2(x)$. 

Poderíamos mostrar o erro também simplesmente colocando o módulo: 

$\displaystyle Err(x) = |f(x) - P_2(x)|$. 

porém o erro não é tão bonito porque ele forma um "bico" em $x_0$. Note a forma de $g(x)$ como ela se parece com um polinômio cúbico. 

```python
x=np.linspace(0.7,1.3,100) #gera um vetor x com 100 valores igualmente espaçados de -2 até 100
cm = 1/2.54  # centimeters in inches
fig, ax = plt.subplots(figsize=(15*cm, 10*cm))
ax.plot(x, f(x) - Poli_Taylor_ord_2(x), label = 'diferença f(x)-P2(x)')
ax.set_xlabel('x',fontsize=14)
ax.set_ylabel('y',fontsize=14)
ax.set_title('Diferença entre f(x) e o poinomio de Taylor',fontsize=14)
ax.legend(loc='lower right') 
ax.grid(True)
plt.show()
```

Os valores abaixo mostram o erro em pontos da forma 

\begin{align}
      x_j &= x_0 + h_j, \\
      h_j &= \frac{0.2}{2^j}
\end{align}
para $j=1,2,\dots,5$. Note que $h_j$ começa valendo $0.1$ e vai sendo dividido por $2$ e por $2$ novamente...

```python
x = x0 + np.array([0.1, 0.05, 0.025, 0.0125, 0.00625]) #np.array transforma numa "lista" só de floats
print('x = ', x)
print('erro = ', np.abs(f(x) - Poli_Taylor_ord_2(x)) )
```
